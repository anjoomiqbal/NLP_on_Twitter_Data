# Step 2 
#### time when users are likely to be on Twitter portal 

### Library Imports

```python
import pandas as pd
import time
import os
```

### Input File

```python
input_file_path = "twitter_data.csv"
df_i = pd.read_csv(input_file_path)
df_i.head() # DEBUG
```

#### drop unnecessary columns for this analysis

```python
df_i.drop(['tweetid', 'month', 'day', 'minute', 'second', 'time_correction', 'year', 'tweet'], axis=1, inplace=True)
df_i.head() # DEBUG
```

### Finiding Unique userids and creating a new output dataframe

```python
unique_users = df_i['userid'].unique()
# print(len(unique_users)) # DEBUG
# print(df_i.columns) # DEBUG
df_i.head() # DEBUG
```

```python
df_o = pd.DataFrame(unique_users, columns=['userid'])
df_o.head() # DEBUG
```

#### creating 8 bins for time of use
night_1     = 00:00 - 02:59<br>
night_2     = 03:00 - 05:59<br>
morning_1   = 06:00 - 08:59<br>
morning_2   = 09:00 - 11:59<br>
afternoon_1 = 12:00 - 14:59<br>
afternoon_2 = 15:00 - 17:59<br>
evening_1   = 18:00 - 20:59<br>
evening_2   = 21:00 - 23:59

#### creating another 7 bins for day of the week
Mon, Tue, Wed, Thu, Fri, Sat, Sun

```python
df_o['total_tweets_by_a_user']=0
df_o['day_mon']=0
df_o['day_tue']=0
df_o['day_wed']=0
df_o['day_thu']=0
df_o['day_fri']=0
df_o['day_sat']=0
df_o['day_sun']=0
df_o['time_night_1']=0
df_o['time_night_2']=0
df_o['time_morning_1']=0
df_o['time_morning_2']=0
df_o['time_afternoon_1']=0
df_o['time_afternoon_2']=0
df_o['time_evening_1']=0
df_o['time_evening_2']=0

# df_o.head() # DEBUG
# df_o.dtypes # DEBUG 
```

```python
start = time.time()

idx_o = 0
idx_i = 0

while idx_i < len(df_i.index):
    
    row_i = df_i.iloc[idx_i]
    row_o = df_o.iloc[idx_o]
    
    # as all the tweets from a single user are all together in the input csv data
    # we can break out of the 2nd for loop to make the program efficient without loosing data
    if row_o['userid'] == row_i['userid']:

        df_o.loc[idx_o, 'total_tweets_by_a_user'] = row_i['total_tweets_by_a_user'] 

        if   row_i['day_of_week'].upper() == 'MON': df_o.loc[idx_o, 'day_mon'] += 1
        elif row_i['day_of_week'].upper() == 'TUE': df_o.loc[idx_o, 'day_tue'] += 1
        elif row_i['day_of_week'].upper() == 'WED': df_o.loc[idx_o, 'day_wed'] += 1
        elif row_i['day_of_week'].upper() == 'THU': df_o.loc[idx_o, 'day_thu'] += 1
        elif row_i['day_of_week'].upper() == 'FRI': df_o.loc[idx_o, 'day_fri'] += 1
        elif row_i['day_of_week'].upper() == 'SAT': df_o.loc[idx_o, 'day_sat'] += 1
        elif row_i['day_of_week'].upper() == 'SUN': df_o.loc[idx_o, 'day_sun'] += 1

        if   row_i['hour'] >= 0 and row_i['hour'] <  3 : df_o.loc[idx_o, 'time_night_1'] += 1
        elif row_i['hour'] >= 3 and row_i['hour'] <  9 : df_o.loc[idx_o, 'time_night_1'] += 1
        elif row_i['hour'] >= 6 and row_i['hour'] <  9 : df_o.loc[idx_o, 'time_morning_1'] += 1
        elif row_i['hour'] >= 9 and row_i['hour'] < 12 : df_o.loc[idx_o, 'time_morning_2'] += 1
        elif row_i['hour'] >=12 and row_i['hour'] < 15 : df_o.loc[idx_o, 'time_afternoon_1'] += 1
        elif row_i['hour'] >=15 and row_i['hour'] < 18 : df_o.loc[idx_o, 'time_afternoon_2'] += 1
        elif row_i['hour'] >=18 and row_i['hour'] < 21 : df_o.loc[idx_o, 'time_evening_1'] += 1
        elif row_i['hour'] >=21 and row_i['hour'] < 24 : df_o.loc[idx_o, 'time_evening_2'] += 1
        else: print ('Time format error: ',row_i['hour'])    
            
    else:
        idx_o += 1      
        idx_i -= 1
        
    idx_i += 1
             
end = time.time()
elapsed = end - start
print(elapsed)
```

## Output to CSV 1 with data

```python
output_file_path = "twitter_user_analytics.csv"

# removing old outout file if exists
if os.path.exists(output_file_path):
    os.remove(output_file_path)

    
# writing to csv
df_o.to_csv(output_file_path, sep=',', index=True, index_label='user_count')
```

#### Normalize the columns 

```python
start = time.time()

for idx_o, row_o in df_o.iterrows():
    x = row_o['total_tweets_by_a_user']
    
    df_o.loc[idx_o, 'day_mon'] /= x
    df_o.loc[idx_o, 'day_tue'] /= x
    df_o.loc[idx_o, 'day_wed'] /= x
    df_o.loc[idx_o, 'day_thu'] /= x
    df_o.loc[idx_o, 'day_fri'] /= x
    df_o.loc[idx_o, 'day_sat'] /= x
    df_o.loc[idx_o, 'day_sun'] /= x

    df_o.loc[idx_o, 'time_night_1'] /= x
    df_o.loc[idx_o, 'time_night_2'] /= x
    df_o.loc[idx_o, 'time_morning_1'] /= x
    df_o.loc[idx_o, 'time_morning_2'] /= x
    df_o.loc[idx_o, 'time_afternoon_1'] /= x
    df_o.loc[idx_o, 'time_afternoon_2'] /= x
    df_o.loc[idx_o, 'time_evening_1'] /= x
    df_o.loc[idx_o, 'time_evening_2'] /= x
                
end = time.time()
elapsed = end - start
print(elapsed)
```

### Output CSV 2 with normalized data

```python
output_file_path = "twitter_user_analytics_normalized.csv"

# removing old outout file if exists
if os.path.exists(output_file_path):
    os.remove(output_file_path)

# writing to csv
df_o.to_csv(output_file_path, sep=',', index=True, index_label='user_count')
```
